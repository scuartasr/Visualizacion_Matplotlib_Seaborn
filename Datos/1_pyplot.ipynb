# Pyplot básico

Para poder hacer un plot básico en Python, la biblioteca más popular es `matplotlib`. El alias habitual de esta biblioteca es `plt`.

```python
# Importación de librerías
from matplotlib import pyplot as plt
import numpy as np
```

Para ejemplificar la realización de un gráfico en Python, se va a crear un conjunto de datos con ayuda de `Numpy`, para graficar de forma discreta a la función $f(x) = x^2$ en el dominio $0 \leq x \leq 5$.

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

Con estos datos definidos, se procede con la graficación de estos datos. Para poderlos graficar, se usa la función `plt.plot(x, y)`, donde `x` corresponde a un vector o un arreglo con los valores de las abscisas, mientras que en `y` se almacena un arreglo con los valores de las ordenadas.

```python
plt.plot(x, y)
plt.show()
```

Nótese que, por defecto, se unen los puntos con una curva suave y se grafica de color azul. Ahora, si se quisiera cambiar el color a uno azul más oscuro, se puede usar a `'b'` como tercer argumento. Veamos:

```python
plt.plot(x, y, 'b')
plt.show()
```

Por otro lado, si en lugar de una línea quisieramos unos puntos rellenos y conservar el color azul, usamos como tercer argumento a `'bo'`.

```python
plt.plot(x, y, 'bo')
plt.show()
```

Ahora, si quisiéramos que en lugar de puntos tuviésemos equis y de color rojo, entonces:

```python
plt.plot(x, y, 'rx')
plt.show()
```

Y para agregarles una línea:

```python
plt.plot(x, y, 'rx-')
plt.show()
```

Se puede observar que con variaciones en el tercer argumento de `plt.plot()` se puede cambiar la estética de nuestro gráfico. En la documentación de `matplotlib` se puede consultar por otras opciones.

Por otro lado, si queremos graficar un histograma, usamos la función `plt.hist()`:

```python
plt.hist(x)
plt.show()
```

Y para un gráfico circular, se puede usar la función `plt.pie()`:

```python
plt.pie(x)
plt.show()
```

Nótese que al entregarle un arreglo de números, le asigna una proporción de tal forma que si el arreglo es:

$x = [x_1, \ x_2, \ \dots, \ x_n]$

Entonces la proporción de área que se tendrá para el $i$-ésimo elemento de $x$, que podemos notar como $A_{x_1}$, y con $i = 1, \ 2, \ \dots, n$, es:

$A_{x_i} = \frac{x_i}{\sum_{k = 1}^n x_k}$

Ahora si se quiere realizar un gráfico de dispersión, se puede usar la función `plt.scatter()`:

```python

plt.scatter(x, y)
plt.show()
```

Para realizar un gráfico de cajas y bigotes:

```python
plt.boxplot(x)
plt.show()
```
