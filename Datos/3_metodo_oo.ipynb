# Método orientado a objeto usando `Matplotlib`.

Además de la subiblioteca `pyplot`, existe una opción orientada a objetos que permite realizar visualizaciones más customizables y estéticas. ¡Es más potente!

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

Para usar el método orientado a objetos, se deben crear dos figuras:
- La primera, típicamente notada como `fig`, puede entenderse como el lienzo sobre el cual se realizarán las diferentes visualizaciones.
- La seguna, típicamente notada como `axes`, permite ubicar los gráficos a realizar en un punto particular. Con este elemento se identificará un gráfico particular.
Veamos:

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x, y, 'b')
plt.show()
```

# Subplots usando el método orientado a objetos

Consideremos un tercer array:

```python
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

Nótese que se crea un lienzo, teniendo gráficos repartidos en una fila y dos columnas.

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2)
ax1.plot(x, y, 'b')
ax2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(2, 3, sharey = True, sharex = True)
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x))
axes[0, 2].plot(x, np.tan(x))
axes[1, 0].plot(x, x)
axes[1, 1].plot(x, -np.abs(x))
axes[1, 2].plot(x, (x + 1))
```
