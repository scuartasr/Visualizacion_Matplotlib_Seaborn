# Elementos estéticos para gráficos con `Seaborn` y `Matplotlib`

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Se comenzará definiendo un par de arreglos con los cuales se realizarán diversos gráficos. Veamos:

```python
x = np.linspace(0, 5, 1000)
y = np.sin(x)
```

Y se graficará un lienzo al cual le iremos agregando elementos estéticos:

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[1].plot(y, x)
```

Como se puede ver, el gráfico no es muy informativo. ¿Qué estámos graficando? ¿Qué representa el eje $y$? ¿Y el eje $x$? Para ello es importante agregar algunos elementos importantes como el título y etiquetas para los gráficos.

```python
fig, axes = plt.subplots(1, 2)
axes[0].plot(x, y)
axes[0].set_title('Relación de X y Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[1].plot(y, x)
axes[1].set_title('Relación Y y X')
```

Como se puede observar, con ayuda del objeto `axes`, puedo acceder a diferentes paneles de nuestro gráfico. Además, con métodos como `set_title`, `set_xlabel` y `set_ylabel` puedo agregar títulos y etiquetas a los ejes. Asimismo, con el argumento `figsize`, al cual se le asigna una tupla como valor, puedo determinar el tamaño del gráfico que quiero realizar, veamos:

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 5))
axes[0].plot(x, y, label = '$y = sin(x)$')
axes[0].set_title('Relación de X y Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y, x, label = '$x = sin(y)$')
axes[1].set_title('Relación Y y X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].legend()
```

Practiquemos con otra gráfica:


```python
plt.plot(x, y, label = '$f(x) = sin(x)$', color = '#e83bee')
plt.title('Gráfica de la función seno')
plt.xlabel('$x$')
plt.ylabel('$y=f(x)$')
plt.legend(loc = 'lower left')
```

# Colores y estilos

Los gráficos pueden ser customizados aún más con el uso de colores y estilos. Con la siguiente instrucción podemos chequear los estilos predefinidos en `matplotlib` para darle un diseño distinto a nuestros gráficos:

```python
print(plt.style.available)
```

Pongámoslo en práctica:

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig, ax = plt.subplots(figsize = (12, 8))
ax.plot(x, x+1, color = 'red', alpha = 0.8, linewidth = 4, linestyle = '-', marker = 'o', markersize = 10)
ax.plot(x, x+2, color = 'blue', alpha = 0.8, linewidth = 1, linestyle = '--', marker = 'x', markersize = 0.5)
ax.plot(x, x+3, color = '#e83bee', alpha = 0.8, linewidth = 2, linestyle = ':', marker = 'v', markersize = 2)
ax.plot(x, x+4, color = 'green', alpha = 0.8, linewidth = 3, linestyle = 'dashed', marker = 'P', markersize = 1.75, markerfacecolor = 'b')
```
