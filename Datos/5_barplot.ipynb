# Gr치fico de barras con `Matplotlib`

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

Comencemos definiendo algunos datos. Como se va a graficar un gr치fico de barras, una de las variables debe ser categ칩rica.

```python
# Varible categ칩rica
country = ['India', 'Japan', 'Mexico', 'Colombia', 'Germany']
# Variable cuantitativa
population = [1500, 180, 140, 51, 82]
```

Para poder graficar estas dos variables, basta con emplear `plt.bar`:

```python
plt.bar(country, population, width = 0.5, color = 'green')
plt.title('Poblaci칩n por pa칤s', fontdict = {'fontsize': 8})
plt.suptitle('Tomando algunos pa칤ses aleatorios')
plt.xticks(np.arange(5), ('India', 'Jap칩n', 'M칠xico', 'Colombia', 'Alemania'),
           rotation = 45)
plt.xlabel('Pa칤s')
plt.ylabel('Poblaci칩n [millones de personas]')
plt.show()
```

Para poder hacer las barras horizontales, basta con cambiar `bar` por `barh`:

```python
plt.barh(country, population, color = 'green')
plt.title('Poblaci칩n por pa칤s')
plt.suptitle('Tomando algunos pa칤ses aleatorios')
plt.yticks(np.arange(5), ('India', 'Jap칩n', 'M칠xico', 'Colombia', 'Alemania'))
plt.ylabel('Pa칤s')
plt.xlabel('Poblaci칩n [millones de personas]')
plt.show()
```
