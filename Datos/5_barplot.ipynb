# Gráfico de barras con `Matplotlib`

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

Comencemos definiendo algunos datos. Como se va a graficar un gráfico de barras, una de las variables debe ser categórica.

```python
# Varible categórica
country = ['India', 'Japan', 'Mexico', 'Colombia', 'Germany']
# Variable cuantitativa
population = [1500, 180, 140, 51, 82]
```

Para poder graficar estas dos variables, basta con emplear `plt.bar`:

```python
plt.bar(country, population, width = 0.5, color = 'green')
plt.title('Población por país', fontdict = {'fontsize': 8})
plt.suptitle('Tomando algunos países aleatorios')
plt.xticks(np.arange(5), ('India', 'Japón', 'México', 'Colombia', 'Alemania'),
           rotation = 45)
plt.xlabel('País')
plt.ylabel('Población [millones de personas]')
plt.show()
```

Para poder hacer las barras horizontales, basta con cambiar `bar` por `barh`:

```python
plt.barh(country, population, color = 'green')
plt.title('Población por país')
plt.suptitle('Tomando algunos países aleatorios')
plt.yticks(np.arange(5), ('India', 'Japón', 'México', 'Colombia', 'Alemania'))
plt.ylabel('País')
plt.xlabel('Población [millones de personas]')
plt.show()
```

## Histograma

Recordar que un histograma permite conocer la distribución asociada a un conjunto de datos numéricos. Veamos entonces el histograma asociado al objeto `datos`:

```python
plt.style.use('seaborn-white')
```

```python
plt.hist(datos, bins = 10, histtype = 'bar')
plt.grid()
plt.show()
```

```python
plt.hist(datos, bins = 10, histtype = 'step')
plt.grid()
plt.show()
```
