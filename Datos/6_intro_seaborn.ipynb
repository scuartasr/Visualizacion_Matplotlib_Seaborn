# `Set` en `Seaborn`

Permite definir un tema, el tipo de fuente y otros elementos de customización.

```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
sns.barplot(x = ['A', 'B', 'C', 'D'],
            y = [10, 7, 9, 12])
plt.show()
```

```python
sns.set(style = 'darkgrid', palette = 'muted', font = 'Segoe UI', font_scale = 1)
sns.barplot(x = ['A', 'B', 'C', 'D'],
            y = [10, 7, 9, 12])
plt.show()
```

# Parámetros más usados con `Seaborn`

Esta biblioteca ofrece una gran cantidad de parámetros que permiten customizar mejor una gráfica.

```python
tips = sns.load_dataset('tips')
```

Estudiemos un marco de datos relacionado con qué tanto se deja de propinas en un restaurante por parte de los clientes.

```python
tips.head()
```

```python
sns.displot(data = tips,
            x = 'total_bill')
plt.title('Histograma para las propinas')
plt.show()
```

```python
sns.displot(data = tips,
            x = 'total_bill',
            hue = 'sex')
plt.show()
```

```python
sns.displot(data = tips,
            x = 'total_bill',
            hue = 'sex',
            kind = 'kde')
plt.show()
```

```python
sns.displot(data = tips,
            x = 'total_bill',
            y = 'tip',
            hue = 'sex')
plt.show()
```

```python
sns.displot(data = tips,
            x = 'total_bill',
            y = 'tip',
            hue = 'sex',
            kind = 'kde')
plt.show()
```
