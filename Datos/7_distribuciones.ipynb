# Distribuciones con `Seaborn`.

`Seaborn` ofrece varios tipos distintos de distribuciones. Vamos a usar los datos `tips` almacenados en `Seaborn`.

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Veamos un histograma para la variable `tip` del marco de datos `tips`:

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12)
plt.show()
```

Ahora, si nos interesa conocer la distribución acumulada, basta con agregar el argumento `cumulative = True`:

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             cumulative = True)
plt.show()
```

Y si nos interesa ver en un mismo gráfica la distribución discriminada por alguna variable categórica, com puede ser `sex`:

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex')
plt.show()
```

Por defecto, el conteo que se refleja en el eye $y$ es absoluto, pero con el argumento `stat = 'probability'` podemos cambiarlo por la frecuencia relativa:

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex',
             stat = 'probability')
plt.show()
```

Aunque si quiero que se muestre en una escala de cero a cien (como un porcentaje), basta con usar el argumento `stat = 'percent`:

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex',
             stat = 'percent')
plt.show()
```

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex',
             stat = 'density')
plt.show()
```

De igual forma, tenemos el argumento `multiple` son el cual podemos elegir la forma en la que queremos presentar los datos:
- `stack`. Apiladas.
- `layer`. Se superponen. Este es el argumento por defecto.
- `dodge`. Barras agrupadas.
- `fill`. Barras con la misma altura para mostrar proporciones.

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex',
             stat = 'count',
             multiple = 'dodge')
plt.show()
```

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex',
             stat = 'count',
             multiple = 'stack')
plt.show()
```

```python
sns.histplot(data = tips,
             x = 'tip',
             bins = 12,
             hue = 'sex',
             stat = 'count',
             multiple = 'fill')
plt.show()
```

Otro gráfico que vale la pena rescatar de `Seaborn` es el `kdeplot`, el cual muestra un gráfico de densidad para una variable. Veámosla para la variable `tip` discriminada según el sexo:

```python
sns.kdeplot(data = tips,
            x = 'tip',
            hue = 'sex')
```

Si nos interesa la densidad acumulada:

```python
sns.kdeplot(data = tips,
            x = 'tip',
            hue = 'sex',
            cumulative = True)
```

O si queremos que haya un relleno suave debajo de cada línea de densidad:

```python
sns.kdeplot(data = tips,
            x = 'tip',
            hue = 'sex',
            cumulative = False,
            shade = True)
```

```python
sns.kdeplot(data = tips,
            x = 'tip',
            hue = 'sex',
            cumulative = False,
            shade = True,
            bw_adjust = 1)
```

Otro gráfico popular es el `ecdfplot`:

```python
sns.ecdfplot(data = tips,
             x = 'tip',
             hue = 'sex')
plt.show()
```

```python
sns.ecdfplot(data = tips,
             x = 'tip',
             hue = 'sex',
             stat = 'count')
plt.show()
```

```python
sns.boxplot(data = tips,
            x = 'sex',
            y = 'tip')
plt.show()
```

```python
sns.displot(data = tips,
            x = 'tip',
            hue = 'sex')
```

```python
sns.displot(data = tips,
            x = 'tip',
            hue = 'sex',
            kind = 'kde')
```

```python
sns.displot(data = tips,
            x = 'tip',
            hue = 'sex',
            kind = 'hist',
            multiple = 'stack')
```
