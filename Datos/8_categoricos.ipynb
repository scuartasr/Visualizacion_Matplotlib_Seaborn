# Gráficos para variables categóricas con `Seaborn`.

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
pd.concat([tips.head(), tips.tail()], axis = 0)
```

```python
sns.set(style = 'darkgrid', palette = 'muted', font = 'Segoe UI', font_scale = 1)
```

```python
sns.countplot(data = tips,
              x = 'day',
              hue = 'sex')
plt.show()
```

```python
sns.countplot(data = tips,
              y = 'day',
              hue = 'sex')
plt.show()
```

```python
sns.stripplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              dodge = True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.swarmplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              dodge = True)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.boxplot(data = tips,
            x = 'day',
            y = 'total_bill',
            hue = 'sex',
            dodge = True)
sns.swarmplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              dodge = True,
              color = '0',
              marker = '<')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.violinplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              dodge = True)
plt.title('Distribución del total de la factura por día según el género')
plt.xticks(np.arange(4), ('Jueves', 'Viernes', 'Sábado', 'Domingo'))
plt.xlabel('Día')
plt.ylabel('Total de la factura [$USD]')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.violinplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              split = True)
plt.title('Distribución del total de la factura por día según el género')
plt.xticks(np.arange(4), ('Jueves', 'Viernes', 'Sábado', 'Domingo'))
plt.xlabel('Día')
plt.ylabel('Total de la factura [$USD]')
plt.legend(loc = 'upper left')
plt.show()
```

```python
sns.catplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              dodge = True,
              kind = 'swarm',
              legend = False)
plt.title('Distribución del total de la factura por día según el género')
plt.xticks(np.arange(4), ('Jueves', 'Viernes', 'Sábado', 'Domingo'))
plt.xlabel('Día')
plt.ylabel('Total de la factura [$USD]')
plt.legend(loc = 'upper left')
plt.show()
plt.show()
```

```python
sns.catplot(data = tips,
              x = 'day',
              y = 'total_bill',
              hue = 'sex',
              dodge = True,
              kind = 'box',
              legend = False,
              col = 'time')
#plt.title('Distribución del total de la factura por día según el género')
plt.xticks(np.arange(4), ('Jueves', 'Viernes', 'Sábado', 'Domingo'))
plt.xlabel('Día')
plt.ylabel('Total de la factura [$USD]')
plt.legend(loc = 'upper left')
plt.show()
plt.show()
```
