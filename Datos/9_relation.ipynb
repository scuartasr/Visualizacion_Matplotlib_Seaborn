# Gráficos de relación en `Seaborn`.

Son gráficos que permiten analizar la correlación entre distintas variables numéricas.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.scatterplot(data = tips, x = 'total_bill', hue = 'day', y = 'tip',
                style = 'time', size = 'size')
plt.legend(loc = 'center', bbox_to_anchor = (1.12, 0.5))
plt.title('Total de la cuenta contra propina dejada')
plt.suptitle('Según el día, el momento y la cantidad de comensales')
plt.show()
```

```python
markers = {'Lunch': 'D', 'Dinner': 's'}
sns.scatterplot(data = tips, x = 'total_bill', hue = 'day', y = 'tip',
                style = 'time', size = 'size', markers = markers)
plt.legend(loc = 'center', bbox_to_anchor = (1.12, 0.5))
plt.title('Total de la cuenta contra propina dejada')
plt.suptitle('Según el día, el momento y la cantidad de comensales')
plt.show()
```

```python
sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
plt.show()
```

```python
sns.relplot(data = tips, x = 'total_bill', hue = 'day', y = 'tip',
                style = 'time', size = 'size', markers = markers)
plt.show()
```
